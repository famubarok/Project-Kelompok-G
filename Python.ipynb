# 1. Persiapan Data
Pada tahap ini kita import library yang dibutuhkan yaitu Pandas.

```python
#import pandas
import pandas as pd
```

Data yang kita gunakan yaitu Daftar Pelabuhan Kargo tersibuk di dunia. Berikut ini merupakan daftar pelabuhan tersibuk berdasar jumlah barang yang dimuat atau diterima melalui dermaga kontainer. Data yang dihitung dalam satuan TEU (twenty-foot equivalent unit) atau standar ISO terhadap satuan kontainer barang.
sumber : https://id.wikipedia.org/wiki/Daftar_pelabuhan_kargo_tersibuk_di_dunia

```python
port_df = pd.read_csv("Pelabuhan.csv", index_col="Peringkat")
for year in range(2006, 2014):
    port_df[str(year)] = port_df[str(year)].convert_dtypes(convert_integer=True)
port_df
```

Selain menampilkan keseluruhan data, kita dapat melihat sebagian data, untuk itu kita gunakan fungsi head(). Secara default pandas akan menampilkan 5 data pertama dari dataframe

```python
# Menampilkan 5 baris teratas
port_df.head()
```

# 2. Pengolahan Data

a. Kita juga dapat melihat 5 pelabuhan tersibuk berdasarkan tahun

```python
# Cek pelabuhan tersibuk berdasarkan tahun
year = int(input("Masukkan tahun (range 2006 s.d. 2013): "))
if year in range(2006, 2014):
    year = str(year)
    
elif year < 2006:
    print("Tahunnya kurang bro!")
else:
    print("Tahunnya kelebihan bro!")

subset = port_df[["Pelabuhan", "Negara", year]]
subset.sort_values(by=[year], ascending=False).head(10)

```
